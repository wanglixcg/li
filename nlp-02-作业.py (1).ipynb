```python

def get_url(url,pattern):
    response=requests.get(url,headers='')
    response.encoding='utf-8'
    pattern_s=re.compile(pattern)
    result=re.findall(pattern_s,response.text)
    return result
```

```python
def get_links():
    url_b='https://baike.baidu.com'+'/item/%E5%8C%97%E4%BA%AC%E5%9C%B0%E9%93%81/408485'
    
    links=get_url(url_b,r'<a target=_blank href="(/item/.+?)">(北京地铁.+?线)</a>')
    links_dic={}
    for herf,name in links:
        links_dic[name]='https://baike.baidu.com'+herf
    return links_dic
        
```

```python
def build_relation():
    station_start_list=dict()
    #print(station_list.values())
    for station in station_list.values():
        #print(station)
        for s in station:
            #print(s)
            if station_start_list.get(s):continue
            station_start_list[s]=station
    
    station_connetion=dict()
    for s in station_start_list.keys():
        #print(s)
        station_connetion[s]=set()
    #print(station_connetion)
   
    for s in  station_start_list.keys():
        #print(s)
        for y in station_start_list.values():
            #print(y)
            
            try:
                index=y.index(s)
                print(index)
                if index == 0:
                    station_connetion[s].add(y[index + 1])
                    #print(y[index+1])
                    #print(station_connetion[s])
                elif index == len(y) - 1:
                    station_connetion[s].add(y[index - 1])
                else:
                    station_connetion[s].add(y[index + 1])
                    station_connetion[s].add(y[index - 1])
            except ValueError:
                continue
           
    return station_connetion
    
```

```python
def bfs(grahp,start,is_goal,search_vis):
    visitied_b=[[start]]
    print(visitied_b)
    seen_b=set()
    while visitied_b:
        visitied_city=visitied_b.pop(0)
        #print('visitied_city is {}'.format(visitied_city))
        fronerier=visitied_city[-1]
        #print('fronerier is {}'.format(fronerier))
        if fronerier in seen_b:continue
        city_list=grahp[fronerier]
        for line in city_list:
            if line in visitied_city:continue
            #print(line)
            path=visitied_city+[line]
            #print(path)
            
            visitied_b.append(path)
            if is_goal in path:return path
            
    seen_b.add(fronerier)
    visitied_b=search_vis(visitied_b)
   
```

```python
def sort_path(cmp_func, beam=-1):
    def _sorted(pathes):
        return sorted(pathes, key=cmp_func)[:beam]
    return _sorted
```
